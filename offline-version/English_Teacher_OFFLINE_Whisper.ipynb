## ğŸ§  2. Load Whisper

```python

import whisper
model = whisper.load_model("base")
print("Whisper loaded")
```

## ğŸ¤ 3. Speech to Text

```python

def transcribe_audio(audio_path):
    result = model.transcribe(audio_path, fp16=False)
    return result["text"]
```

## ğŸ“ 4. Offline Teacher Logic

```python

class OfflineEnglishTeacher:
    def detect_level(self, text):
        if len(text.split()) < 5:
            return "beginner"
        elif len(text.split()) < 12:
            return "intermediate"
        else:
            return "advanced"

    def lesson(self, level):
        lessons = {
            "beginner": "Repeat: I am learning English. What is your name?",
            "intermediate": "Great! Tell me what you did yesterday.",
            "advanced": "Excellent! Discuss the impact of technology."
        }
        return lessons[level]
```

## ğŸ”Š 5. Text to Speech

```python
import pyttsx3

# Install espeak-ng as it's required by pyttsx3
!apt-get update && apt-get install -y espeak-ng

engine = pyttsx3.init()
engine.setProperty("rate", 160)

def speak(text):
    engine.say(text)
    engine.runAndWait()
```

## ğŸš€ 6. Run Demo

```python

teacher = OfflineEnglishTeacher()

text = input("Type something in English: ")
level = teacher.detect_level(text)

print("Detected level:", level)
response = teacher.lesson(level)

print("Teacher:", response)
speak(response)
```
